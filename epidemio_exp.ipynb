```python
from benchmark.epidemiology.simulation import Simulation
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime as dt

simulation = Simulation()
```

```python
time_duration = 351
start_date = dt.datetime.strptime("2020-01-03", "%Y-%m-%d")
days = [start_date + dt.timedelta(days=i) for i in range(time_duration)]
""" then = start_date + dt.timedelta(days=time_duration)
days = mdates.drange(start_date, then, dt.timedelta(days=1)) """
```

### French government

```python
params = [0.5, 0.7, 0.5]

res = simulation(params, get_all_results=True)

plt.figure(figsize=(10,6))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%B %Y'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=31))
plt.plot(days, res)
plt.xlabel("time")
plt.ylabel("Cum. nb deaths")
plt.grid(True, linestyle="--")
plt.gcf().autofmt_xdate()
plt.savefig("french_gov.svg", bbox_inches="tight")
```

```python
max_deaths = res[-1]
```

### AdaLIPO

```python
params = [0.33300191, 0.66948696, 0.99413936]

res = simulation(params, get_all_results=True)

plt.figure(figsize=(10,6))
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%B %Y'))
plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=31))
plt.plot(res)
plt.xlabel("time")
plt.ylabel("Cum. nb deaths")
plt.ylim(0, max_deaths)
plt.grid(True, linestyle="--")
plt.gcf().autofmt_xdate()
plt.savefig("AdaLIPO.svg", bbox_inches="tight")
```
